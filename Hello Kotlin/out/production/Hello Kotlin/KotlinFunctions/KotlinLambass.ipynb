```kotlin
//Practice Time
val spices = listOf("curry", "pepper", "cayenne", "ginger", "red curry", "green curry", "red pepper")

val curries = spices.filter { it.contains("curry") }.sortedBy { it.length }
println(curries)

val cSpices1 = spices.filter { it.startsWith('c') && it.endsWith('e') }
println(cSpices1)

val cSpices2 = spices.filter { it.startsWith('c') }.filter { it.endsWith('e') }
println(cSpices2)

val firstThreeWithC = spices.take(3).filter { it.startsWith('c') }
println(firstThreeWithC)
```
